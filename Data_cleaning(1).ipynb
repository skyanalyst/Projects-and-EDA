# Step 2 : To remove null vlaues

```python
# delete null values using drop na 
# first way
df_1.dropna()
```

```python
# Second way
df_1.dropna(how='all')
```

```python
# fill na values
# method 1 : to fill with 0
df_1.fillna(0)
```

```python
# method 2: forward fill with 'ffill'
df_1.fillna(method='ffill')
```

```python
# method_2: fill with backward fill usinf 'bfill'
df_1.fillna(method='bfill')
```

```python
df_1.info()
```

```python
# Drop columns with 0 non null values
df_1.drop(['Street Number Suffix','TIDF Compliance','Site Permit','Voluntary Soft-Story Retrofit','Structural Notification','Unit Suffix'],axis=1)
```

```python
df_1
```

```python
# method_3 : fill with interpolate
df_dropped=df_1.interpolate()
```
